```python
from qgis.PyQt.QtWidgets import QApplication, QMainWindow, QTableView , QVBoxLayout, QCheckBox, QWidget, QHBoxLayout
from qgis.PyQt.QtCore import Qt
import sys
from qgis.PyQt.QtGui import QStandardItemModel, QStandardItem

class MainWindow(QMainWindow):
    def __init__(self):
        super().__init__()

        self.initUI()

    def initUI(self):
        # 创建主布局
        main_layout = QVBoxLayout()

        # 创建模型和视图
        self.model = QStandardItemModel(self)
        self.model.setHorizontalHeaderLabels(["名称", "类别", "尺寸", "数量", "选择"])
        self.tableView = QTableView()
        self.tableView.setModel(self.model)
        self.populateModel()

        # 创建表头复选框（实际上是一个单独的复选框，放在表格上方）
        self.headerCheckBox = QCheckBox("选择所有")
        self.headerCheckBox.stateChanged.connect(self.onHeaderCheckBoxStateChanged)

        # 创建布局来容纳复选框和表格视图
        checkbox_layout = QHBoxLayout()  # 这个布局实际上在这个例子中是多余的，因为只有一个复选框
        checkbox_layout.addWidget(self.headerCheckBox)

        # 将复选框布局和表格视图添加到主布局中
        main_layout.addLayout(checkbox_layout)  # 这将复选框放在表格上方
        main_layout.addWidget(self.tableView)

        # 创建一个中心窗口小部件，并将布局设置给它
        central_widget = QWidget()
        central_widget.setLayout(main_layout)
        self.setCentralWidget(central_widget)

    def populateModel(self):
        # 填充模型数据，包括复选框项
        for i in range(5):
            row = [
                QStandardItem(f"名称 {i+1}"),
                QStandardItem(f"类别 {i+1}"),
                QStandardItem(f"尺寸 {i+1}"),
                QStandardItem(f"数量 {i+1}"),
                QStandardItem()  # 复选框列，初始为空，稍后设置
            ]
            checkbox_item = row[-1]
            checkbox_item.setCheckable(True)
            checkbox_item.setCheckState(Qt.Unchecked)
            self.model.appendRow(row)

    def onHeaderCheckBoxStateChanged(self, state):
        # 根据表头复选框的状态更新表格中所有行的复选框项
        for row in range(self.model.rowCount()):
            checkbox_item = self.model.item(row, 4)  # 复选框列在第五列（索引为4）
            if checkbox_item.isCheckable():
                checkbox_item.setCheckState(state)

if __name__ == '__main__':
    app = QApplication(sys.argv)
    window = MainWindow()
    window.show()
    sys.exit(app.exec_())
```
